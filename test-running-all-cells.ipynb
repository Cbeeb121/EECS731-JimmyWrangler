```python
population_df = pd.read_csv('data/raw/population_by_country_2020.csv')
```

```python
df['mean'] = df.mean(axis = 1)
```

```python
df.sort_values(by="mean", ascending=False)
```

```python
gross_population_suicide_rate_df = pd.merge(population_df[["Population (2020)","Country"]], df, on='Country')
```

```python
gross_population_suicide_rate_df
```

```python
gross_population_suicide_rate_df['average_number_of_suicides'] = (gross_population_suicide_rate_df["Population (2020)"] / 100000.0) * gross_population_suicide_rate_df["mean"]
```

```python
gross_population_suicide_rate_df
```

```python
population_vs_average_suicides = gross_population_suicide_rate_df[gross_population_suicide_rate_df['Sex'].isin([' Both sexes'])]
```

```python
top_10_countries_by_suicide = population_vs_average_suicides[["Country", "average_number_of_suicides", "Population (2020)"]].sort_values(by="average_number_of_suicides", ascending=False).head(10)
```

```python
top_10_countries_by_suicide.plot(x="Country", y= ["average_number_of_suicides"], kind="bar")
```

```python
gross_population_suicide_rate_df['suicide rate vs population'] = gross_population_suicide_rate_df["average_number_of_suicides"] / gross_population_suicide_rate_df["Population (2020)"]
```

```python
gross_population_suicide_rate_df
```

```python
top_10_countries_by_suicide.sort_values("Population (2020)", ascending=False).plot(x="Country", y= "Population (2020)", kind="bar")
```
